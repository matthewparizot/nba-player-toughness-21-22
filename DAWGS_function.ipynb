#### High DAWG Actions
Offensive Rebound - O
Drawing Offensive Fouls - D
And1 - O
LooseBallRecovered OFF/DEF O-D
Charges - D

#### Medium DAWG Actions
Defensive Rebound - D
Steal - D
Block - D
Deflection - D
Offensive Box Out - O

#### Low DAWG Actions
Drawing Shooting Fouls (if not AND1) - O
Defensive Box Out - D
Making Contest Shots - O
Screen Assists - O

#### Negative DAWG Actions
Turnovers (Lost Ball) - O

```python
# Functions to calculate Offensive DAWG, Defensive DAWG

def OFFdawgs(ix):
    ORB = 0.3 * ix[6]
    SA = 0.05 * ix[16]
    CON = ((ix[21] + ix[22]) * 0.1) * ix[5]
    OBO = (0.25 * ix[23]) * (ix[25]/100)
    SFD = (ix[13] - ix[15]) * 0.1
    AND1 = ix[15] * 1.25
    OLBR = ix[18] * 1.25
    TOLB = ix[12] * 0.1

    OFFdawg_score = ORB + SA + CON + OBO + SFD + AND1 + OLBR - TOLB

    return OFFdawg_score 

def DEFdawgs(ix):
    DRB = 0.1 * ix[7]
    STL = 0.25 * ix[9]
    BLK = 0.25 * ix[10]
    OFD = (ix[14] - ix[20]) * 0.5
    DEFL = 0.1 * ix[17]
    DLBR = ix[19] * 1.25
    CHAR = ix[20] * 1.25
    DBO = (0.1 * ix[24]) * (ix[25]/100)

    DEFdawg_score = DRB + STL + BLK + OFD + DEFL + DLBR + CHAR + DBO

    return DEFdawg_score

# Function to create new columns in dataframes

def dawg_score(df):
    off_dawg_score = []
    def_dawg_score = []
    total_dawg_score = []
    off_dawg_usage = []
    def_dawg_usage = []
    tot_dawg_usage = []

    for n in range(len(df)):
        player = df.iloc[n]

        ODS = OFFdawgs(player)
        DDS = DEFdawgs(player)

        off_dawg_score.append(ODS)
        def_dawg_score.append(DDS)
        total_dawg_score.append(ODS + DDS)
        off_dawg_usage.append(ODS/(player[4]/player[3]))
        def_dawg_usage.append(DDS/(player[4]/player[3]))
        tot_dawg_usage.append((ODS+DDS)/(player[4]/player[3]))
    
    df['OffDAWG'] = pd.Series(off_dawg_score)
    df['DefDAWG'] = pd.Series(def_dawg_score)
    df['TotDAWG'] = pd.Series(total_dawg_score)
    df['OffDAWGusage'] = pd.Series(off_dawg_usage)
    df['DefDAWGusage'] = pd.Series(def_dawg_usage)
    df['TotDAWGusage'] = pd.Series(tot_dawg_usage)
```

```python
dawg_score(bulls)
```
